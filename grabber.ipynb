```python
import pybullet as p
import time
import pybullet_data
import numpy as n
import context_retriever_main as llm


final_res = 0

def full_script(input_question):#llm코드 실행행
    global final_res
    
    for i in range(3):#llm 재대로 답하면 한번에 break 이상한답은 새번 다시 물어보기
        answer = llm.LLM_chatbot(input_question)
        if llm.evaluator(answer):
            print(f"answer is ",answer)
            final_res = answer
            break
        
        else:
            print("no proper response:")
            print(answer+"\n")

question = input("tell me, what should i point at?")
full_script(question)#simulation 실행전 먼저 llm질문과답실행

###################################### pybullet 로봇 실행 및 세팅    #######################
physicsClient = p.connect(p.GUI)#or p.DIRECT for non-graphical version
p.setAdditionalSearchPath(pybullet_data.getDataPath()) #optionally
p.resetSimulation()
p.setGravity(0,0,-9.8)
p.setRealTimeSimulation(0)
planeId = p.loadURDF("plane.urdf")

startPos = [0,0,0]
startOrientation = p.getQuaternionFromEuler([0,0,1])
#xarm = p.loadURDF("xarm/xarm6_with_gripper.urdf",startPos ,startOrientation ,useFixedBase = 1)
kuka = p.loadURDF("kuka_experimental-melodic-devel/kuka_lbr_iiwa_support/urdf/lbr_iiwa_14_r820.urdf",startPos ,startOrientation ,useFixedBase = 1)
objects = p.loadSDF("gripper/wsg50_one_motor_gripper_new_free_base.sdf")
gripper = objects[0]

teddy = p.loadURDF("teddy_vhacd.urdf", 1.050000, -0.500000, 0.700000, 0.000000, 0.000000, 0.707107, 0.707107)
cube = p.loadURDF("cube_small.urdf", 0.950000, -0.100000, 0.700000, 0.000000, 0.000000, 0.707107, 0.707107)
sphere = p.loadURDF("sphere_small.urdf", 0.850000, -0.400000, 0.700000, 0.000000, 0.000000, 0.707107,0.707107)
duck = p.loadURDF("duck_vhacd.urdf", 0.850000, -0.400000, 0.900000, 0.000000, 0.000000, 0.707107,0.707107)


Orientation = p.getQuaternionFromEuler([3.14,0.,0.])

#print("kuka joints: {}".format(p.getNumJoints(kuka)))
#########################################################################################################


def arm_control(result):
  duckPos, duckOrn = p.getBasePositionAndOrientation(duck)
  teddyPos, teddyOrn = p.getBasePositionAndOrientation(teddy)
  #cubePos, cubeOrn = p.getBasePositionAndOrientation(cube)
  spherePos, sphereOrn = p.getBasePositionAndOrientation(sphere)
  #[sphere, duck, bear, none]
  
  if result == "sphere":
    target_pos = p.calculateInverseKinematics(kuka, 7, spherePos,targetOrientation = Orientation)#targetOrientation = Orientation
    p.setJointMotorControlArray(kuka, range(7), p.POSITION_CONTROL, targetPositions=target_pos) 
  elif result == "duck":
    target_pos = p.calculateInverseKinematics(kuka, 7, duckPos,targetOrientation = Orientation)#targetOrientation = Orientation
    p.setJointMotorControlArray(kuka, range(7), p.POSITION_CONTROL, targetPositions=target_pos)
  elif result == "bear":
    target_pos = p.calculateInverseKinematics(kuka, 7, teddyPos,targetOrientation = Orientation)#targetOrientation = Orientation
    p.setJointMotorControlArray(kuka, range(7), p.POSITION_CONTROL, targetPositions=target_pos)  
  elif result == "none":
    target_pos = p.calculateInverseKinematics(kuka, 7, [0,0,0],targetOrientation = Orientation)#targetOrientation = Orientation
    p.setJointMotorControlArray(kuka, range(7), p.POSITION_CONTROL, targetPositions=target_pos)  
  else:
    print("no input")

while(1):
    print("################final: ",final_res)
    #keyboard_control() #화살표로 로봇 운행 testing
    arm_control(final_res)
    p.stepSimulation()
    time.sleep(1./240.)
p.disconnect()
```
